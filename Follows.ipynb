```r
# Importing libraries for data visualization
library(ggplot2)
library(dplyr)
library(tidyr)
library(lubridate)

#Import the necessary libraries
library(ggplot2)
```

```r

# Reading the csv file into a data frame
instagram_data <- read.csv("insta_data.csv")



# Finding the number of null values in each column
colSums(is.na(instagram_data))

# Dropping rows with null values
instagram_data <- na.omit(instagram_data)
```

```r

# Calculate the total number of follows
total_follows <- sum(instagram_data$Follows)

# Display the result
cat("The total number of follows from posts are:", total_follows)

sorted_data <- instagram_data[order(instagram_data$Follows, decreasing = TRUE),]
```

```r
# Calculate the average number of follows per post
avg_follows_per_post <- sum(instagram_data$Follows) / nrow(instagram_data)

# Display the result
cat("The average number of follows per post is:", avg_follows_per_post)
```

```r
# Create a scatterplot of follows from home
ggplot(instagram_data, aes(x = Follows, y = From.Home)) +
  geom_point() +
  labs(title = "Follows from Home",
       x = "Follows",
       y = "From Home") +
  theme_minimal()
```

```r
# Create a scatterplot of follows from hashtags
ggplot(instagram_data, aes(x = Follows, y = From.Hashtags)) +
  geom_point() +
  labs(title = "Follows from Hashtags",
       x = "Follows",
       y = "From Hashtags") +
  theme_minimal()
```

```r
# Create a scatterplot of follows from explore
ggplot(instagram_data, aes(x = Follows, y = From.Explore)) +
  geom_point() +
  labs(title = "Follows from Explore",
       x = "From Explore",
       y = "Follows") +
  theme_minimal()
```
