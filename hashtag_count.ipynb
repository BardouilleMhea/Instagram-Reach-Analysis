```r

# Reading the csv file into a data frame
instagram_data <- read.csv("insta_data.csv")



# Finding the number of null values in each column
colSums(is.na(instagram_data))

# Dropping rows with null values
instagram_data <- na.omit(instagram_data)
```

```r
# Create a new dataframe with the count of each hashtag
hashtags <- instagram_data %>%
  separate_rows(Hashtags, sep = ",") %>%
  group_by(Hashtags) %>%
  summarise(count = n()) %>%
  arrange(desc(count))

# Print the top 10 hashtags by count
head(hashtags, 10)
```
